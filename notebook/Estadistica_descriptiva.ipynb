## Importacion de librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import seaborn as sns
```

```python
# Cargar los datos
data = pd.read_csv('../data/most-spotify-2024.csv', encoding='latin1')

data_clean = pd.read_csv('../data/data_limpia.csv', encoding='latin1')
```

## Graficas con datos limpios

### Distribucion de streamings por plataforma

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['Spotify Streams'] / 1_000_000, bins=30)
plt.title("Distribución de Streams en Spotify (en millones)")
plt.xlabel("Spotify Streams (millones)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['YouTube Views'] / 1_000_000, bins=30)
plt.title("Distribución de Visualizaciones en YouTube (en millones)")
plt.xlabel("YouTube Views (millones)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['Shazam Counts'] / 1_000_000, bins=30)
plt.title("Distribución de Búsquedas en Shazam (en millones)")
plt.xlabel("Shazam Counts (millones)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['Track Score'], bins=30)
plt.title("Distribución del Track Score")
plt.xlabel("Track Score")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#Cantindad canciones por año de lanzamiento
plt.figure(figsize=(10,5)) 
sns.histplot(data_clean["Release Year"], bins=30, kde=False)
```

```python
data_clean['Artist'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Artistas con más Canciones en el Dataset")
plt.show()
```

```python
sns.pairplot(data_clean[['Spotify Streams','YouTube Views','TikTok Views','Track Score']])
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data_clean.corr(numeric_only=True), cmap="coolwarm")
plt.title("Matriz de Correlación - Datos Limpios")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=data_clean['Spotify Streams'] / 1_000_000)
plt.title("Boxplot de Spotify Streams (en millones)")
plt.xlabel("Spotify Streams (millones)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=data_clean['YouTube Views'] / 1_000_000)
plt.title("Boxplot de YouTube Views (en millones)")
plt.xlabel("YouTube Views (millones)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=data_clean['TikTok Views'] / 1_000_000)
plt.title("Boxplot de TikTok Views (en millones)")
plt.xlabel("TikTok Views (millones)")
plt.show()
```
