Estos errores hubieran afectado directamente el rendimiento del modelo de clasificación elegido, especialmente porque nuestra variable objetivo, Explicit Track, es categórica y requiere un tratamiento adecuado para que el modelo pueda aprender correctamente. Por esto, fue necesario realizar una limpieza y estandarización previa antes de continuar con la construcción del modelo.

## Graficas con datos limpios

### Distribucion de streamings por plataforma

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['Spotify Streams'] / 1_000_000, bins=30)
plt.title("Distribución de Streams en Spotify (en millones)")
plt.xlabel("Spotify Streams (millones)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['YouTube Views'] / 1_000_000, bins=30)
plt.title("Distribución de Visualizaciones en YouTube (en millones)")
plt.xlabel("YouTube Views (millones)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['Shazam Counts'] / 1_000_000, bins=30)
plt.title("Distribución de Búsquedas en Shazam (en millones)")
plt.xlabel("Shazam Counts (millones)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data_clean['Track Score'], bins=30)
plt.title("Distribución del Track Score")
plt.xlabel("Track Score")
plt.ylabel("Frecuencia")
plt.show()
```

```python
#Cantindad canciones por año de lanzamiento
plt.figure(figsize=(10,5)) 
sns.histplot(data_clean["Release Year"], bins=30, kde=False)
```

Esta grafica de relacion de los años de exposicion de canciones son mas recientes, se puede deducir que empezo a crecer desde la misma epoca en que aparecen las plataformas y su crecimiento en popularidad, otra observacion que se debe tener en cuenta es que las canciones que se lanzaron antes de esas caracteristicas no son comparables con la popularidad que obtuvieron las que son contemporaneas con la plataforma.
Se podria proyectar incluso que aunque la base de datos no lo incluye, puede tener relacion proporcional que la popularida de una cancion haga popular a la plataforma y le genera mas visitas a la plataforma.

```python
data_clean['Artist'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Artistas con más Canciones en el Dataset")
plt.show()
```

Una observacion para agregar es posible que la visibilidad de un artista potencia la popularidad de la cancion, esto quiere decir que pueden existir datos atipicos de cuando una cancion tiene artistas no muy reconocidos y en contra parte cuando varias canciones que entran en el ranking de las mas escuchadas pertenecen al mismo artista en la base de datos puede generar duplicados.
